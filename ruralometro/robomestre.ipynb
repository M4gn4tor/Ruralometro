# Calcula estatísticas do grupo febril, acima ou igual 37,5 - cada variável de filtro é um indicador

```python
grupo_febril = geral[(geral['TEMPERATURA'] > 37.5) | (geral['TEMPERATURA'] == 37.5) ]

filtro1 = grupo_febril[(grupo_febril['Politico_Bancada_ruralista'] == 'Sim')]

filtro2 = grupo_febril[(grupo_febril['Politico_Frente_Ambientalista'] == 'Sim')]

filtro3 = grupo_febril[(grupo_febril['Tem_doacao_Ibama'] == 'Sim')]

filtro4 = grupo_febril[(grupo_febril['Tem_doacao_TE'] == 'Sim')]

filtro5 = grupo_febril[(grupo_febril['Tem_divida_INSS'] == 'Sim')]

filtro6 = grupo_febril[(grupo_febril['Tem_Funrural'] == 'Sim')]

votosfebril = pd.merge(grupo_febril, votos, left_on='CPF_candidato', right_on='Politico_CPF', how = 'left')

filtro7 = votosfebril[(votosfebril['Projeto_Voto_y'] == 'Sim') & (votosfebril['Projeto_Relevancia_y'] == 'Muito relevante') & (votosfebril['Projeto_Avaliacao_y'] == 'Desfavorável')]

filtro8 = votosfebril[(votosfebril['Projeto_Voto_y'] == 'Não') & (votosfebril['Projeto_Relevancia_y'] == 'Muito relevante') & (votosfebril['Projeto_Avaliacao_y'] == 'Desfavorável')]

projetosfebril = pd.merge(grupo_febril, projetos, left_on='CPF_candidato', right_on='CPF', how = 'left')

filtro9 = projetosfebril[(projetosfebril['Qual_avaliacao'] == 'Desfavorável')]

filtro10 = projetosfebril[(projetosfebril['Qual_avaliacao'] == 'Favorável')]
```
